```python
# 兼容 pythone2,3
from __future__ import print_function

# 导入相关python库
import os

#使用matplotlib库画图
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn
from sklearn import datasets

#设定随机数种子
np.random.seed(36)
```

```python
#读取数据
housing = pd.read_csv('./dateset/美国KingCounty的房屋销售价格/kc_train.csv')
target = pd.read_csv('./dateset/美国KingCounty的房屋销售价格/target.csv')  #销售价格
t = pd.read_csv('./dateset/美国KingCounty的房屋销售价格/kc_test.csv')      #测试数据
```

```python
#数据预处理
housing.info()    #查看是否有缺失值
```

```python
#特征缩放
from sklearn.preprocessing import MinMaxScaler  ## 归一化到[0,1]
minmax_scaler = MinMaxScaler()
minmax_scaler.fit(housing)   #进行内部拟合，内部参数会发生变化
scaler_housing = minmax_scaler.transform(housing)
scaler_housing = pd.DataFrame(scaler_housing,columns=housing.columns)
scaler_housing[:5]
```

```python
housing[:5]
```

```python
mm = MinMaxScaler()
mm.fit(t)
scaler_t = mm.transform(t)
scaler_t = pd.DataFrame(scaler_t,columns=t.columns)
scaler_t[:5]
```

```python
#选择基于梯度下降的线性回归模型
from sklearn.linear_model import LinearRegression
LR_reg = LinearRegression()
#进行拟合
LR_reg.fit(scaler_housing,target)
```

```python
#使用均方误差用于评价模型好坏
from sklearn.metrics import mean_squared_error
preds = LR_reg.predict(scaler_housing)   #输入数据进行预测得到结果
mse = mean_squared_error(preds,target)   #使用均方误差来评价模型好坏，可以输出mse进行查看评价值
mse
```

```python
#绘图进行比较
plt.figure(figsize=(12,8))       #画布大小
num = 500
x = np.arange(1,num+1)              #取100个点进行比较
plt.plot(x,target[:num],label='target')      #目标取值
plt.plot(x,preds[:num],label='preds')        #预测取值
plt.grid()
plt.legend(loc='upper right')  #线条显示位置
plt.show()
```

```python
#输出测试数据
result = LR_reg.predict(scaler_t)
df_result = pd.DataFrame(result)
df_result.to_csv("./dateset/美国KingCounty的房屋销售价格/result.csv")
result[:5]
```
